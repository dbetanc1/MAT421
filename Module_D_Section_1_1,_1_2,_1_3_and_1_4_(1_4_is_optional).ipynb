## 1.2.3 Gram-Schmidt Process

The Gram-Schmidt algorithm is used to obtain an orthonormal basis. Let
a1,...,am be linearly independent. We intend to find an orthonormal basis of
span(a1,...,am). The process takes advantage of the properties of the orthogo-
nal projection derived above. In essence we add the vectors ai one by one, but
only after taking out their orthogonal projection on the previously included
vectors. The outcome spans the same subspace and orthogonal decomposition
ensures orthogonality.
